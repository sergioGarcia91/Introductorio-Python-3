# Introducción
Matplotlib es una potente librería de Python diseñada para la creación de gráficos en dos dimensiones. Su conjunto de herramientas robustas se concentra principalmente en el módulo `matplotlib.pyplot`, que facilita la creación de visualizaciones. Algunas de sus funcionalidades a destacar son:

- [`plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html): Crear gráficos de líneas.
- [`scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html): Crear gráficos de dispersión.
- [`bar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html): Crear gráficos de barras verticales.
- [`barh`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html): Crear gráficos de barras horizontales.
- [`hist`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html): Crear histogramas.
- [`pie`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html): Crear gráficos de pastel.
- [`boxplot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html): Crear gráficos de caja.
- [`contour`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html): Crear gráficos de contorno.
- [`savefig`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html): Guardar el gráfico en un archivo.

Estas son solo algunas de las funcionalidades que Matplotlib ofrece. Además, cuenta con [tutoriales](https://matplotlib.org/stable/tutorials/index) y [ejemplos](https://matplotlib.org/stable/gallery/index.html) en su [documentación oficial](https://matplotlib.org/stable/tutorials/index.html).

Estos recursos son una guía desde los conceptos fundamentales hasta conceptos más avanzados para la visualización de datos en Python.

La claridad y la autoexplicación son fundamentales en la creación de gráficos.

````python
plt.figure(figsize=(10, 5))

# Gráfico de la función seno
plt.plot(x, # Valores en el eje X
         y, # Valores en el eje Y
         label='Función Seno' # Etiqueta o label para referencia la leyenda
         )

# Configuración del gráfico
plt.xlabel('Eje X') # Etiqueta del eje X
plt.ylabel('Eje Y') # Etiqueta del eje Y
plt.title('Gráfico de Líneas') # Título de la gráfica
plt.legend() # Activamos que aparezca la leyenda si tenemos un label para la gráfica

plt.show()
````

Matplotlib ofrece una amplia variedad de colores que se pueden utilizar para personalizar los gráficos. Se puede consultar la lista completa de colores en la [documentación de colores](https://matplotlib.org/stable/gallery/color/named_colors.html).


````python
plt.figure(figsize=(10, 5))

plt.plot(x,
         y,
         label='Función Seno',
         color='tab:green', # Especificar el color a utilizar
         )

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')
plt.legend()

plt.show()
````

El parámetro `linestyle` o `ls`en la función `plot` de Matplotlib se utiliza para especificar el estilo de línea que se debe utilizar en el gráfico. Puedes elegir entre diferentes estilos de línea según tus preferencias.

Aquí hay algunos ejemplos comunes de estilos de línea que puedes utilizar:

- `'-'`: Línea sólida (predeterminada).
- `'--'`: Línea punteada.
- `':'`: Línea punteada más densa.
- `'-.'`: Línea punteada y sólida.

````python
plt.figure(figsize=(10, 5))

plt.plot(x,
         y,
         label='Función Seno',
         color='tab:green',
         linestyle='--' # Cambiar el estilo de la línea, tambien se puede llamar como ls
         )

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')
plt.legend()

plt.show()
````

Se pueden incluir múltiples gráficos en un mismo lienzo simplemente añadiendo más líneas de código para cada gráfico.

````python
# Usemos los mismos valores de x
# pero creemos un nuevo y2
y2 = np.sin(2*x)

# y un y3
y3 = 2*x + 1
````

````python
plt.figure(figsize=(10, 5))

# Grafico de y
plt.plot(x,
         y,
         label='$f(x)=sin(x)$', # con formato Latex
         color='tab:green',
         linestyle='--'
         )

# Gráfico de y2
plt.plot(x,
         y2,
         label='f(x)=sin(2x)', # sin formato Latex
         color='red',
         linestyle='-.'
         )

# Gráfico de y3
plt.plot(x,
         y3,
         label='$f(x)=2x+1$', # con formato Latex
         )


plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')
plt.legend()

plt.show()
````

Se puede ajustar manualmente los límites de los ejes x e y utilizando las funciones `plt.xlim()` y `plt.ylim()` en Matplotlib.

````python
plt.figure(figsize=(6, 5))

# Grafico de y
plt.plot(x, y, label='$f(x)=sin(x)$', color='tab:green', linestyle='--')

# Gráfico de y2
plt.plot(x, y2, label='f(x)=sin(2x)', color='red', linestyle='-.' )

# Gráfico de y3
plt.plot(x, y3, label='$f(x)=2x+1$')

plt.xlim(-3, 6) # Ajuste limites eje X (valor mínimo, valor mximo)
plt.ylim(-0.25, 5) # Ajuste limites eje I (valor mínimo, valor mximo)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')
plt.legend()
plt.grid(ls='--', color='k') # Tambien se puede editar la grilla

plt.show()
````

# Histograma

````python
# Usaremos la funcion numpyp.random.randn()
# para generar dos grupos de datos con distribución normal
muestra_1 = np.random.randn(1000)
muestra_2 = np.random.randn(250) + 4

# Unir los dos arrays
datos = np.concatenate((muestra_1, muestra_2))

plt.figure()

plt.hist(datos,
         bins=20,        # La cantidad de celdas o barras a considerar
         color='c',   # Color de las barras
         edgecolor='k'  # Color del borde de las barras
         )

plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma')

plt.show()
````

````python
plt.figure()

plt.hist(datos/3, bins=20, color='cyan', edgecolor='black', label='Datos / 3')
plt.hist(datos, bins=100, alpha=0.7, label='Datos')
# alpha es un parámetro para controlar la transparencia de los gráficos

plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.legend()

plt.show()
````

# Gráfico de Dispersión

````python
# Crearemos una nube puntos que consideren la
# ecuacion de la recta:
# y = mx + b
# donde m = 2.5 , b = -10
# x entre 10.5 y 90.5
x = np.arange(10.5, 90.5, 3.25)
m = 2.5
b = -10
y = m*x+b

#vamos a considerar algo de ruido sumado a los datos de y
ruido = (100*np.random.rand( len(y) )) - 50
y_con_ruido = y + ruido # ambos deben tener la misma longitud

# Gráfico de y sin ruido
plt.scatter(x, y,
            color='k',
            marker='o',
            s= 20, # para el tamaño de los puntos
            label='$y=2.5 \cdot x - 10$',
            alpha= 0.5) # transparencia

# Gráfico de y con ruido
plt.scatter(x, y_con_ruido,
            color='red',
            marker='x',
            label='$y$ con ruido')

# También podemos incluir un gráfico de línea
plt.plot(x, y, ls=':')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

plt.show()
````

Si se ejecuta la celda anterior varias veces, se obtendrá salidas diferentes cada vez. Esto es debido a que los números se generan aleatoriamente en cada ejecución. Para asegurar la reproducibilidad y obtener siempre los mismos números aleatorios, se puede hacer uso de una semilla `seed`. Para ello se utiliza la función `numpy.random.seed()`.

Ejemplo:
```python
# Establecer la semilla para la reproducibilidad
np.random.seed(seed=2)

ruido = (100 * np.random.rand(len(y))) - 50
```
Al establecer una semilla específica, en este caso de 2, los números aleatorios generado en la ejecucion posterior serán siempre los mismos.

````python
x = np.arange(10.5, 90.5, 3.25)
m = 2.5
b = -10
y = m*x+b

# Establecer la semilla para la reproducibilidad
np.random.seed(seed=2)

ruido = (100*np.random.rand( len(y) )) - 50
y_con_ruido = y + ruido

plt.scatter(x, y, color='k', marker='o', s= 20,
            label='$y=2.5 \cdot x - 10$', alpha= 0.5)

plt.scatter(x, y_con_ruido, color='red',
            marker='x', label='$y$ con ruido')

plt.plot(x, y, ls=':')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

plt.show()
````
